# Create dataset

This notebook pretends to create an analyse the main csv file and dataset.

Starts by checking for duplicates and creating the raw dataset

The preprocessment is applied in order to create final_dataset.csv and add it to data/ folder

### Initial setup

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from auxiliary.auxiliary_funcs import create_merged_dataset

csv_files_path = 'csv_files/'
output_file_raw_dataset ="../data/final_dataset.csv"
```

# Fazer dataset completo

```python
df_all_data = create_merged_dataset(csv_files_path, output_file_raw_dataset)
df_all_data.drop_duplicates(subset=['local'], inplace=True)
df_all_data.sort_values(by='local', inplace=True)
df_all_data.to_csv(output_file_raw_dataset, index=False)
#df_all_data.to_excel('dataset_teste.xlsx', index=False)  # For debugging purposes
```

```python
# Set up the plotting environment
plt.figure(figsize=(20, 15))


# Check for unwanted duplicates
plt.figure(figsize=(12, 6))
df_all_data['local'].value_counts().head(100).plot(kind='bar', color='orange', edgecolor='black')
plt.title('Frequency of appeareances (should all be 1.0)', fontsize=14)
plt.xlabel('Local')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()

# Histogram for numerical columns
df_all_data.hist(bins=20, figsize=(20, 15), color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Numerical Columns', fontsize=16)
plt.show()
```
